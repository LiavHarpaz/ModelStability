```python
from typing import List

class SameResults:
    """
    Measure and log the rows in `left` that appear in `right`,
    the rows are equal base on `columns_to_measure`.
    The 100% is the num of rows in `left`.

    Parameters
    ----------
    same_row_cols : List
        List of columns that should be the same in both dataframes to indicate a same row.
    percent_value: bool, default False
    log_name : str, default 'same_results'
        The string that used for the msg/message fields in the logger

    Examples
    --------
    same_results = SameResults(['id1','id2','grade1','grade2','rank1','rank2'])
    same_results(left, right)
    
    Returns
    -------
    int or float
        The value of the same results.
    """

    def __init__(self, same_row_cols: List, percent_value: bool = False, log_name : str = 'same_results') -> None:
        self.same_row_cols = same_row_cols
        self.percent_value = percent_value
        self.log_name = log_name


    def __call__(self, left, right) -> int or float:
#         TODO: replace with minus and count()
        results = len(pd.merge(left,right,on=self.same_row_cols))
        if self.percent_value:
            results = results/len(left)

        print(f'{self.log_name} '
             f'statistic_type: percent_{self.percent_value} '
             f'statistic_value:{results}')
        return results
```

```python
from typing import List

class ResultsOnlyInLeft:
    """
    Measure and log the num of rows in `left` that doesnt show up in `right`.

    Parameters
    ----------
    same_row_cols : List
        List of columns that should be the same in both dataframes to indicate a same row.
    percent_value: bool, default False
    should_distinct : bool
        Should the dataframe be distinct (distinct by `cols`)
    log_name : str
        The string that used for the msg/message fields in the logger

    Examples
    --------
    results_added = ResultsOnlyInLeft(['id1','id2'], False, 'results_added')
    results_added(new,old)
    
    Returns
    -------
    int
        Num of results that only show up in the left dataframe.
    """
    def __init__(self, same_row_cols: List, should_distinct: bool ,log_msg: str) -> None:
        self.same_row_cols = same_row_cols
        self.should_distinct = should_distinct
        self.log_msg = log_msg
        
    def __call__(self, left, right)-> int:
        left = left[self.same_row_cols]
        right = right[self.same_row_cols]
        if self.should_distinct:
            left = left.drop_duplicates()
            right = right.drop_duplicates()
        results_in_left =  self.subtract_dataframes(left,right,self.same_row_cols)
        results_in_left_cnt = len(results_in_left)

        print(f'{self.log_msg} '
             f'statistic_value:{results_in_left_cnt} '
              f'ids:{self.same_row_cols} ')
        return results_in_left_cnt
        
    @staticmethod
    def subtract_dataframes(left,right,cols):
        #TODO: replace with spark's minus
        results =  pd.merge(
                    left, right,
                    how='outer', on=cols, indicator=True, suffixes=('_foo','')).query(
                        '_merge == "left_only"')
        return results
```

```python
import math
from typing import List

class CompareColumnGrade:
    """
    Measure and log the average diff in grades in the `left` dataframe compared to `right` dataframe
    above a threshold: > 0.9, > 0.8 ...

    Parameters
    ----------
    same_row_cols : List
        List of columns that should be the same in both dataframes to indicate a same row.
    grade_cols : List
        List of grade columns to compare
    log_name : str, default 'grades_shift_above_threshold'
        The string that used for the msg/message fields in the logger

    Examples
    --------
    grades_comparator = CompareColumnGrade(['id1','id2'], ['grade1','grade2'])
    grades_comparator(old,new)
    
    Returns
    -------
    dict
        dict where the keys is the threshold and the value is the diff with each col in `grade_cols` having own sub-dict.
    """
    def __init__(self, same_row_cols: List, grade_cols: List, log_name: str ='grades_shift_above_threshold') -> None:
        self.same_row_cols = same_row_cols
        self.grade_cols = grade_cols
        self.log_name = log_name
        
    def __call__(self,left,right) -> dict:
        left = left.copy()
        right = right.copy()
        results_dict = {}
        for grade_col in self.grade_cols:
            results_dict[grade_col] = {}
            left_grade_col = f'{grade_col}_left'
            right_grade_col = f'{grade_col}_right'
            grade_diff_col = f'{grade_col}_diff'
            
            left=left.rename({grade_col:left_grade_col},axis = 'columns')
            right=right.rename({grade_col:right_grade_col},axis = 'columns')
            both_df=pd.merge(left,right,on=self.same_row_cols)
            
            
            for x in range(10):
                both_df = both_df.copy()
                #not changing the col value in order to keep accuracy
                both_df = both_df[both_df[left_grade_col]>=(x/10)]
                both_df[grade_diff_col] = both_df[right_grade_col] - both_df[left_grade_col]
                diff = both_df[grade_diff_col].mean()
                results_dict[grade_col][(x/10)] = diff
                print(f'{self.log_name} '
                     f'statistic_type: {(x/10)} '
                     f'statistic_value:{diff} '
                      f'id:{grade_col} ')
        return results_dict
```

```python
from typing import List

class StabilityByRank:
    """
    Measure and log the percent of rows in each rank that stayed the same

    Parameters
    ----------
    rank_cols : List
        List of rank columns
    columns_to_compare : List
        List of columns to compate their values
    log_name : str, default 'grades_shift_above_threshold'
        The string that used for the msg/message fields in the logger

    Examples
    --------
    ranks_stability = StabilityByRank(['rank1','rank2'],['id1','id2','rank1','rank2','grade1','grade2'])
    ranks_stability(old,new)
    
    Returns
    -------
    dict
        dict where the keys is the threshold and the value is the stability with each col in `rank_cols` having own sub-dict.
    """
    def __init__(self, rank_cols: List, columns_to_compare: List, log_name: str = 'stability_by_rank') -> None:
        self.rank_cols = rank_cols
        self.columns_to_compare = columns_to_compare
        self.log_name = log_name
        
    def __call__(self,left,right) -> dict:
        left = left.copy()
        right = right.copy()
        results_dict = {}
        for rank_col in self.rank_cols:
            results_dict[rank_col] = {}
            max_rank = left[rank_col].max()
            for rank in range(1,int(max_rank)+1):
                left_rank = left[left[rank_col]==rank]
                rank_count = len(left_rank)
                same_result =pd.merge(left_rank,right,on=self.columns_to_compare)
                same_results_count = len(same_result)
                result = same_results_count/rank_count
                results_dict[rank_col][rank] = result
                print(f'{self.log_name} '
                      f'threshold: {rank} '
                     f'statistic_value:{result} '
                      f'id:{rank_col} ')
        return results_dict
```

```python
import math
from typing import List

class RowsInGradeGroup:
    """
    Measure and log how many rows are in each grade group (0.0-0.1, 0.1-0.2, ...)

    Parameters
    ----------
    grade_cols : List
        List of grade columns
    log_name : str, default 'rows_in_grade_group'
        The string that used for the msg/message fields in the logger

    Examples
    --------
    grades_count = RowsInGradeGroup(['grade1', 'grade2'])
    grades_count(new)
    
    Returns
    -------
    dict
        dict where the keys is the threshold and the value is the num of rows 
        with each col in `grade_cols` having own sub-dict.
    """
    
    def __init__(self, grade_cols: List, log_name: str = 'rows_in_grade_group') -> None:
        self.grade_cols = grade_cols
        self.log_name = log_name
        
    def __call__(self,df) -> dict:
        df = df.copy()
        results_dict = {}
        for grade_col in self.grade_cols:
            results_dict[grade_col] = {}
            df[grade_col] = df[grade_col].apply(lambda x: math.floor(x*10)/10)
            #TODO: groupby, count, collect and for loop
            grade_groups = df.groupby(grade_col).count().to_dict()
            random_key = list(grade_groups.keys())[0]
            for k,v in grade_groups[random_key].items():
                results_dict[grade_col][k] = v
                print(f'{self.log_name} '
                      f'threshold: {k} '
                      f'statistic_value:{v} '
                      f'id: {grade_col} ')
        return results_dict
```

```python
class GradesBetweenRanks:
    """
    Measure and log the average grade diff between ranks.

    Parameters
    ----------
    !! The parameters need to be ordered the same way
    id_cols : List
        List of id columns
    rank_cols : List
        List of rank columns
    grade_cols : List
        List of grade columns
    log_name : str, default 'ranks_grade_diff'
        The string that used for the msg/message fields in the logger

    Examples
    --------
    rank_grade_diff = GradesBetweenRanks(['id1','id2'],['rank_for_id1','rank_for_id2'],['grade_for_id1','grade_for_id2'])
    rank_grade_diff(new)
    
    Returns
    -------
    dict
        dict where the keys is the rank and the value is the diff between the rank and rank+1
        with each col in `id_cols` having own sub-dict.
    """
    
    def __init__(self, id_cols, rank_cols, grade_cols, log_name = 'ranks_grade_diff'):
        self.id_cols = id_cols
        self.rank_cols = rank_cols
        self.grade_cols = grade_cols
        self.log_name = log_name
        
    def __call__(self,df):
        df = df.copy()
        results_dict = {}
        for id_col, rank_col, grade_col in zip(self.id_cols, self.rank_cols, self.grade_cols):
            results_dict[id_col] = {}
            #TODO: replace with window function
            previous_ranks = df.rename({grade_col:'pre_grade',rank_col:'rank'},axis = 'columns')
            previous_ranks['rank'] = previous_ranks['rank'] - 1
            previous_ranks = previous_ranks[previous_ranks['rank'] > 0]
            new_rank = df.rename({rank_col:'rank'},axis = 'columns')
            previous_ranks = pd.merge(new_rank,previous_ranks,on=[id_col,'rank'])
            #end of TODO
            
            previous_ranks['grade_diff'] =  previous_ranks[grade_col] - previous_ranks['pre_grade']
            previous_ranks = previous_ranks.groupby('rank').mean()['grade_diff'].to_dict()
            for k,v in previous_ranks.items():
                results_dict[id_col][k] = v
                print(f'{self.log_name} '
                      f'statistic_type: average '
                      f'threshold: {k} '
                      f'statistic_value:{v} '
                      f'id:{id_col} ')
            
        return results_dict
```
